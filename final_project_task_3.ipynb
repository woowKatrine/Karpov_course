**Задание 3**

В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. 

1) **С помощью каких метрик можно оценить результаты последнего прошедшего события?**

Вопрос, для чего проводятся тематические события. Предположим, что задача была:
- повысить retention на долгосрочной перспективе- всегда хорошо. Метрика- n-day retention. Тестирование на разных группах пользователей.   
- увеличить время, проведённое в игре. Особенно если мы зарабатываем на показе рекламы. Метрика - ASD — Average Session Duration - среднее время сессии за день
- увеличить количество покупок ARPU/ARPPU Средний доход с одного пользователя/платящего пользователя. 
- Конверсия в покупку. Если предположить, что покупка чаще всего совершается на n-й день в игре и с помощью тем.события мы удерживаем игрока и доводим его до нужного дня. 


2) **Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад.**
- Изменится ли набор метрик оценки результата? 

Опять же, вопрос- для чего, какой результат мы хотим увидеть? Предположим- мы хотим замедлить прохождение игры, чтобы сэкономить на разработке новых уровней, увеличить вовлечённость (акцент в игре не на быстрое прохождение, а на более вдумчивое, с получением ачивок и прокачкой)? Или увеличить время, проведённое в игре, потому, что мы зарабатываем на показе рекламы раз в n минут и продаже "отключения рекламы"? 

При любом изменении важно 
1. Не потерять в качестве других метрик
2. Улучшить целевые метрики. 

В первую очередь надо убедиться, что не просядут бизнес-метрики, т.е. те, которые связаны с прибылью
- **retention** 
- **ARPPU** - средний доход на платящего пользователя
- **ARPU** - средний доход на привлеченного пользователя
- **LTV**
- **Время, проведённое в игре**

Целевой метрикой тут может быть
- Время, проведённое в игре и retention, если у нас такая бизнес-модель. Время должно увеличиться, возвращаемость то же
- Если предусмотрена возможность "откупиться от отката", то LTV должно увеличиться, при условии, что игрокам зайдет новая модель игры.
